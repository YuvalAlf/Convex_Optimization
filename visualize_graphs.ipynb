```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
random_df = pd.read_csv('random.csv')
random_df['algorithm'] = 'random'
incremental_df = pd.read_csv('incremental.csv')
incremental_df['algorithm'] = 'incremental'
```

```python
results = pd.concat([random_df, incremental_df], ignore_index=True)
```

```python
results[results['Basis Size'] == 20].sort_values(['algorithm', 'Error'])
```

```python
sns.scatterplot(data=results, x='Basis Size', y='Error', hue='algorithm')
```

```python
plt.figure(figsize=[10, 10])
sns.scatterplot(data=results, x='Basis Size', y='Error', hue='algorithm')
```

```python
plt.figure(figsize=[10, 10])
sns.regplot(data=results, x='Basis Size', y='Error', hue='algorithm')
```
